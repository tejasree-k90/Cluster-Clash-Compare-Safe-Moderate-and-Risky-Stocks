```python
import pandas as pd

# Load the Excel workbook
price_xls = pd.ExcelFile("D:\\D Drive\\Woxsen\\Projects\\esg and finance\\sp500_price_data1.xlsx")

# Dictionary to hold the long-format DataFrames for each sheet
long_dfs = {}

# Process each sheet in the workbook
for sheet in price_xls.sheet_names:
    # Read the current sheet into a DataFrame
    df = pd.read_excel(price_xls, sheet_name=sheet)
    
    # Convert to long format using "Date" as the identifier and the sheet name as the value column name
    price_data_long = df.melt(id_vars=["Date"], var_name="Symbol", value_name=sheet)
    
    # Save the long DataFrame in the dictionary
    long_dfs[sheet] = price_data_long

# Example: print the first few rows of the long-format DataFrame for the 'Close' sheet
print(long_dfs["Close"].head())
```

```python
long_dfs
```

```python
for sheet, df in long_dfs.items():
    null_count = df.isnull().sum()
    print(f"Null values in sheet '{sheet}':")
    print(null_count)
    print("-" * 50)
```

```python
#dropping null values in the sheets
for sheet, df in long_dfs.items():
    long_dfs[sheet] = df.dropna()
```

```python
for sheet, df in long_dfs.items():
    null_count = df.isnull().sum()
    print(f"Null values in sheet '{sheet}':")
    print(null_count)
    print("-" * 50)
```

```python
esg_df.columns
```

```python
merged_prices = None

for sheet, df in long_dfs.items():
    if merged_prices is None:
        merged_prices = df
    else:
        # Merge on Date and Symbol so that each price type becomes its own column
        merged_prices = pd.merge(merged_prices, df, on=["Date", "Symbol"], how="outer")

# Now merge the combined price data with esg_df on "Symbol"
final_merged_data = pd.merge(merged_prices, esg_df, on="Symbol", how="inner")

# Check the result
print(final_merged_data.head())
```
